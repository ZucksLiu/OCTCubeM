```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
import shutil
```

```python
name_list = pd.read_csv('name_list.csv')
name_list
```

```python
save_dir = 'cls_images/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

os.makedirs(save_dir + 'stage2/', exist_ok=True)
os.makedirs(save_dir + 'stage3/', exist_ok=True)
os.makedirs(save_dir + 'stage4/', exist_ok=True)
raw_img_dir = 'Images/'
processed_img_dir = 'Processed_Images/'
for row, values in name_list.iterrows():
    stage = values['stage']
    name = values['name']
    pat_id = values['pat_id']
    eye_id = values['eye_id']
    raw_list = os.listdir(raw_img_dir + str(eye_id))
    filtered_list = [f for f in raw_list if f.endswith('.png')]
    num_frames = len(filtered_list)

    print(stage, name, pat_id, eye_id, num_frames)
    file_list = os.listdir(processed_img_dir + str(eye_id))
    oct_files = sorted([f for f in file_list if f.startswith('oct')], key=lambda x: int(x.split('-')[1].split('.')[0]))

    if stage == 1:
        continue
    assert len(oct_files) == num_frames
    for i, f in enumerate(oct_files):
        oct_img_path = processed_img_dir + str(eye_id) + '/' + f
        frame_idx = int(f.split('-')[1].split('.')[0])

        save_img_path = save_dir + 'stage' + str(stage) + '/' + name + f'_{frame_idx}' + '.png'

        # cp oct_img_path to save_img_path, use shutil
        shutil.copy(oct_img_path, save_img_path)

    print(f'finish {name} {pat_id} {eye_id} {stage} {num_frames} {frame_idx} {oct_img_path} {save_img_path}\n')
```
